```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('Data\\ASCII\\Xeek_train_subset_clean.csv').tail(1000)
df.head()
```

```python
# Split data into training and testing sets (70% training, 30% testing)
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)

train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
```

```python
X_train = train_df[['DEPTH_MD','CALI', 'RDEP', 'RHOB', 'GR', 'NPHI', 'PEF']].fillna(0)  # Fill NaN values with 0
y_train = train_df['DTC'].fillna(0)
X_train.head()
```

```python
y_train.head()
```

```python
X_test = test_df[['DEPTH_MD','CALI', 'RDEP', 'RHOB', 'GR', 'NPHI', 'PEF']].fillna(0)
y_test = test_df['DTC'].fillna(0)
y_test.head()
```

```python
# Linear Regression model
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred_lin = lin_reg.predict(X_test)
lin_mse = mean_squared_error(y_test, y_pred_lin)
print(f"Linear Regression MSE: {lin_mse}")
```

```python
plt.figure(figsize=(6, 30))
plt.plot(y_test, y_test.index, label='Actual DTC', color='blue')
plt.plot(y_pred_lin, y_test.index, label='Predicted DTC', color='red', linestyle='--')
plt.gca().invert_yaxis()  # Invert the depth axis
plt.xlabel('DTC')
plt.ylabel('Depth (MD)')
plt.title('Oilfield Log: Predicted vs Actual DTC')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
rf_mse = mean_squared_error(y_test, y_pred_rf)
print(f"Random Forest MSE: {rf_mse}")
```

```python
plt.figure(figsize=(6, 30))
plt.plot(y_test, y_test.index, label='Actual DTC', color='blue')
plt.plot(y_pred_lin, y_test.index, label='Linear Predicted DTC', color='red', linestyle='--')
plt.plot(y_pred_rf, y_test.index, label='Random Forrest Predicted DTC', color='green', linestyle='--')
plt.gca().invert_yaxis()  # Invert the depth axis
plt.xlabel('DTC')
plt.ylabel('Depth (MD)')
plt.title('Oilfield Log: Predicted vs Actual DTC')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gradient Boosting model
gb = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
gb_mse = mean_squared_error(y_test, y_pred_gb)
print(f"Gradient Boosting MSE: {gb_mse}")
```

```python
plt.figure(figsize=(6, 30))
plt.plot(y_test, y_test.index, label='Actual DTC', color='blue')
plt.plot(y_pred_lin, y_test.index, label='Linear Predicted DTC', color='red', linestyle='--')
plt.plot(y_pred_rf, y_test.index, label='Random Forrest Predicted DTC', color='green', linestyle='--')
plt.plot(y_pred_gb, y_test.index, label='Gradient Boosting Predicted DTC', color='lime', linestyle='--')
plt.gca().invert_yaxis()  # Invert the depth axis
plt.xlabel('DTC')
plt.ylabel('Depth (MD)')
plt.title('Oilfield Log: Predicted vs Actual DTC')
plt.legend()
plt.grid(True)
plt.show()
```

```python
pred_df=test_df
pred_df.head()
```

```python
pred_df['Pred DTC lin']=y_pred_lin
pred_df['Pred DTC rf']=y_pred_rf
pred_df['Pred DTC gb']=y_pred_gb
pred_df.head()
```

```python
pred_df = pred_df.set_index('DEPTH_MD',drop=True)
pred_df.head()
```

```python
pred_df = pred_df.sort_index(ascending=True)
pred_df.head()
```

```python
pred_df = pred_df.reset_index(drop=False)
pred_df.head()
```

```python
import matplotlib.pyplot as plt

# Set up the figure and axes
Fig, ax = plt.subplots(figsize=(15, 10))

# Set up the plot axes
ax1 = plt.subplot2grid((1, 4), (0, 0), rowspan=1, colspan=1)  # Gamma Ray track
ax2 = plt.subplot2grid((1, 4), (0, 1), rowspan=1, colspan=1)  # Density track
ax3 = plt.subplot2grid((1, 4), (0, 2), rowspan=1, colspan=1)  # Neutron track
ax4 = plt.subplot2grid((1, 4), (0, 3), rowspan=1, colspan=1)  # Sonic track for DTC and predicted curves

# Gamma Ray track
ax1.plot(pred_df["GR"], pred_df.index, color="green")  # Plot GR against depth
ax1.set_xlabel("Gamma Ray")
ax1.xaxis.label.set_color("green")
ax1.set_xlim(0, 200)
ax1.set_ylabel("Depth (m)")
ax1.tick_params(axis='x', colors="green")
ax1.spines["top"].set_edgecolor("green")
ax1.set_ylim(pred_df.index.max(), pred_df.index.min())  # Invert depth axis
ax1.grid(which='major', color='lightgrey', linestyle='-')

# Density track
ax2.plot(pred_df["RHOB"], pred_df.index, color="red")  # Plot RHOB against depth
ax2.set_xlabel("Density (RHOB)")
ax2.set_xlim(1.95, 2.95)
ax2.xaxis.label.set_color("red")
ax2.tick_params(axis='x', colors="red")
ax2.spines["top"].set_edgecolor("red")
ax2.set_ylim(pred_df.index.max(), pred_df.index.min())  # Invert depth axis
ax2.grid(which='major', color='lightgrey', linestyle='-')

# Neutron track
ax3.plot(pred_df["NPHI"], pred_df.index, color="blue")  # Plot NPHI against depth
ax3.set_xlabel("Neutron (NPHI)")
ax3.set_xlim(0.45, -0.15)
ax3.xaxis.label.set_color("blue")
ax3.tick_params(axis='x', colors="blue")
ax3.spines["top"].set_edgecolor("blue")
ax3.set_ylim(pred_df.index.max(), pred_df.index.min())  # Invert depth axis
ax3.grid(which='major', color='lightgrey', linestyle='-')

# Sonic track (for DTC and predicted DTC curves)
ax4.plot(pred_df["DTC"], pred_df.index, label='DTC', color="purple")  # Plot DTC against depth
ax4.plot(pred_df["Pred DTC lin"], pred_df.index, label='Pred DTC lin', color="orange", linestyle='--')
ax4.plot(pred_df["Pred DTC rf"], pred_df.index, label='Pred DTC rf', color="cyan", linestyle='--')
ax4.plot(pred_df["Pred DTC gb"], pred_df.index, label='Pred DTC gb', color="magenta", linestyle='--')
ax4.set_xlabel("Sonic (DTC)")
ax4.xaxis.label.set_color("purple")
ax4.set_xlim(140, 40)  # Reverse scale for DTC
ax4.tick_params(axis='x', colors="purple")
ax4.spines["top"].set_edgecolor("purple")
ax4.set_ylim(pred_df.index.max(), pred_df.index.min())  # Invert depth axis
ax4.grid(which='major', color='lightgrey', linestyle='-')
ax4.legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
